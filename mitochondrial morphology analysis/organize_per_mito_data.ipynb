```python
import pandas as pd
```

```python
from pathlib import Path

# List of uploaded CSV files
csv_files = [
    'R2_3D Analysis Data - per Mito.csv',
    'R1-3D Analysis Data - per Mito.csv',
    'L2_3D Analysis Data - per Mito.csv',
    'L1-3D Analysis Data - per Mito.csv'
]

# Initialize an empty list to hold DataFrames
dfs = []

# Base path where the files are stored
base_path = '/Users/yuanyuan/Downloads/200nM deepred/d10_n/d10_n_9-81/L1-L4 deconvolutionized/selected_slices/results_1.45_cval_5/'

# Loop through each CSV file to read and filter data
for csv_file in csv_files:
    # Construct the full path to the CSV file
    full_path = Path(base_path) / csv_file
    
    # Read the CSV file into a DataFrame
    df = pd.read_csv(full_path)
    
    # Filter rows based on 'Volume' column
    df_filtered = df[df['Volume'] >= 0.6]
    
    # Add an additional column for experiment identifier
    experiment_id = Path(csv_file).stem  # Use the file name (minus extension) as the experiment ID
    df_filtered['Experiment ID'] = experiment_id
    
    # Append the filtered DataFrame to the list
    dfs.append(df_filtered)

# Combine all the filtered DataFrames into a single master DataFrame
master_df = pd.concat(dfs, ignore_index=True)


# Eliminate the 'DeepRed Intensity' column if it exists in the DataFrame
if 'deepred Intensity' in master_df.columns:
    master_df.drop('deepred Intensity', axis=1, inplace=True)

# Add a new column named 'Treatment', initializing with a default value (e.g., 'Unknown')
master_df['Treatment'] = 'n_1'

# Show the first few rows of the updated DataFrame to confirm the changes
master_df.head()



# Show first few rows of the master DataFrame
master_df.head()
```

```python
##calculate median Sphericity
# Calculate the median of 'Sphericity' for each experiment, grouped by 'Experiment ID'
median_sphericity_by_experiment = master_df.groupby('Experiment ID')['Sphericity'].median().sort_index()
median_sphericity_by_experiment
```
